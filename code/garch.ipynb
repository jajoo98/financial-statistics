```python
import warnings
warnings.simplefilter(action='ignore')
from arch import arch_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data and set index to datetime in the dataframe
data = pd.read_table("data/nasdaq_w_logret.txt")
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date', drop=True)
alpha, beta, omega = [], [], []
data = data
```

```python
data.describe()
```

```python
data['logreturn'].plot(figsize=(20,5));
plt.title("Nasdaq weekly log returns");
```

### Part (a)

First, we fit a GARCH(1, 1) model with Gaussian innovations to the entire series.

```python
model = arch_model(data, mean='Zero', vol='GARCH', p=1, q=1)
model_fit = model.fit(disp='off')
print(model_fit.summary())
omega.append(model_fit.params[0])
alpha.append(model_fit.params[1])
beta.append(model_fit.params[2])
```

### Part (b)
Now we divide the sample period into four parts: November 19, 1984 to
June 16, 1987; June 9, 1987 to August 15, 1990; August 8, 1990 to March 13, 1998; and March 6, 1998 to September 15, 2003 and fit a GARCH(1, 1) model with Gaussian innovations to the NASDAQ weekly log returns during each period.

```python
divider = data.index.get_loc('1987-06-15')
data1 = data.iloc[:divider+1,:]
divider2 = data.index.get_loc('1990-08-13')
data2 = data.iloc[divider-1:divider2+1,:]
divider3 = data.index.get_loc('1998-03-16')
data3 = data.iloc[divider2-1:divider3,:]
data4 = data.iloc[divider3-2:,:]
```

```python
model1 = arch_model(data1, mean='Zero', vol='GARCH', p=1, q=1)
model1_fit = model1.fit(disp='off')
print(model1_fit.summary())
omega.append(model1_fit.params[0])
alpha.append(model1_fit.params[1])
beta.append(model1_fit.params[2])
```

```python
model2 = arch_model(data2, mean='Zero', vol='GARCH', p=1, q=1)
model2_fit = model2.fit(disp='off')
print(model2_fit.summary())
omega.append(model2_fit.params[0])
alpha.append(model2_fit.params[1])
beta.append(model2_fit.params[2])
```

```python
model3 = arch_model(data3, mean='Zero', vol='GARCH', p=1, q=1)
model3_fit = model3.fit(disp='off')
print(model3_fit.summary())
omega.append(model3_fit.params[0])
alpha.append(model3_fit.params[1])
beta.append(model3_fit.params[2])
```

```python
model4 = arch_model(data4, mean='Zero', vol='GARCH', p=1, q=1)
model4_fit = model4.fit(disp='off')
print(model4_fit.summary())
omega.append(model4_fit.params[0])
alpha.append(model4_fit.params[1])
beta.append(model4_fit.params[2])
```

### Part (c)
Here we compare the results in (a) and (b), and discuss the possibility of parameter changes and their implications on volatility persistence.

```python
ab = [alpha[i]+beta[i] for i in range(5)]
halflife = [-np.log(2)/np.log(i) for i in ab]
table = pd.DataFrame({'omega':omega, 'alpha':alpha, 'beta':beta, 'alpha+beta':ab, 'half-life':halflife})
table = table.set_index(pd.Index(['Full model','Period 1','Period 2','Period 3','Period 4']))
table
```

$\alpha + \beta$ represents volatility persistence. The closer the value to 1, the higher the volatility persistence. For the entire model, and for periods 2, 3 and 4, we can see that the value is quite close to 1. This indicates that the volatility shock would last for a longer time (equivalently indicated by a higher value of half life). For period 1, the volatility persistence is lower, resulting from a near 0 value of $\alpha$ and a relatively low $\beta$. 

The dependence of the log return today depends on past return with a coefficient $\alpha$ and on past volatility with coefficient $\beta$. Hence for period 1, the dependence on the past is low, and equivalently it has low half life and low volatility persistence.

The parameters through the various periods vary a bit - both $\alpha$ and $\beta$. As mentioned before, a lower $\alpha$ indicates lower dependence on past returns, which is seen in periods 1 and 3 specifically. Period 2 has a low value for $\beta$ and period 3 is quite high, indicating low and high dependence on past volatility. The variation of parameters throughout the periods tell us that dividing the full model into smaller periods can give us better insights into the properties of the series at different times.
