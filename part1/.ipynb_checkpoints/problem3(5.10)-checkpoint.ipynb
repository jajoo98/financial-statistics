```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import statsmodels.api as sm
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
mkt_data = pd.read_table("m_sp500ret_3mtcm.txt")
stock_data = pd.read_table("m_logret_10stocks.txt")
```

```python
mkt_data.reset_index(inplace=True)
mkt_data.columns = mkt_data.iloc[0]
mkt_data.dropna(axis=1, inplace=True)
mkt_data.drop(0, inplace=True)
mkt_data['Date'] = pd.to_datetime(mkt_data['Date'], format='%b-%y')
mkt_data = mkt_data.set_index('Date')
mkt_data['sp500'] = (mkt_data['sp500']).astype(float) + 1
mkt_data['3mTCM'] = (mkt_data['3mTCM']).astype(float)*(0.01) + 1
mkt_data = mkt_data.apply(np.log)
mkt_data['3mTCM'] = mkt_data['3mTCM']*(1/12)
#mkt_data
```

```python
stock_data = stock_data.dropna(axis=0)
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.set_index('Date', inplace= True)
stock_data.index += pd.offsets.MonthBegin()
stock_data.index = stock_data.index.shift(-1, freq='MS')
#stock_data
```

```python
divider = mkt_data.index.get_loc('1998-06-01')
stock_data_1 = stock_data.iloc[:divider+1,:]
stock_data_2 = stock_data.iloc[divider+1:,:]
mkt_data_1 = mkt_data.iloc[:divider+1,:]
mkt_data_2 = mkt_data.iloc[divider+1:,:]
#mkt_data_1.shape
```

```python
def capm(stock_ret, mkt_ret, rf):
    excess_ret = stock_ret - rf
    erp = mkt_ret - rf
    model = sm.OLS(excess_ret, erp)
    results = model.fit()
    return results.params[0], 0, results.bse[0], results.mse_resid
```

```python
beta1, alpha1, stde1, v_resid = [], [], [], []
beta2, alpha2, stde2 = [], [], []
for i in range(stock_data_1.shape[1]):
    b, a, stde, v = capm(stock_data_1.iloc[:,i], mkt_data_1.loc[:,'sp500'], mkt_data_1.loc[:,'3mTCM'])
    beta1.append(b)
    alpha1.append(a)
    stde1.append(stde)
    v_resid.append(v)
for i in range(stock_data_2.shape[1]):
    b, a, stde, ignore = capm(stock_data_2.iloc[:,i], mkt_data_2.loc[:,'sp500'], mkt_data_2.loc[:,'3mTCM'])
    beta2.append(b)
    alpha2.append(a)
    stde2.append(stde)
```

```python
x = np.linspace(0,100,stock_data.shape[1])
my_xticks = stock_data.columns
plt.xticks(x, my_xticks);
plt.scatter(x, beta1, label='old betas')
plt.scatter(x, beta2, label='new betas')
plt.xlabel("Stock name");
plt.ylabel("beta for different periods");
fig2 = plt.gcf()
fig2.set_size_inches(8, 5)
plt.legend(loc="upper right");
```

```python
z_values = {}
for i in range(len(beta1)):
    z = (beta1[i]-beta2[i])/(math.sqrt((stde1[i]**2)+(stde2[i]**2)))
    z_values[stock_data.columns[i]] = (z, abs(z)>1.96)
z_values
```

We can test whether the beta values have significantly changed using these z-values. For a 95% significance level, the value of the z-stat is 1.96. While we just reject the null for AMD, a few others are quite high as well and we should have a look at them.

```python
# kalman filtering

beta = beta1 # starting point
beta_list = []
sigma_w_sq = 0.04
sigma_sq = np.square(stde1)
p = np.square(stde1) + sigma_w_sq
#print(p)
for i in range(stock_data_2.shape[0]):
    h = mkt_data_2.iloc[i,0] - mkt_data_2.iloc[i,1]
    y = stock_data_2.iloc[i,:] - mkt_data_2.iloc[i,1]
    kg_term1 = np.multiply(p,h)
    kg_term2 = np.multiply(kg_term1, h) + v_resid
    kalman_gain = np.divide(kg_term1, kg_term2)
    beta = beta + np.multiply(kalman_gain, y-np.multiply(h,beta))
    p = np.multiply(1-np.multiply(kalman_gain,h),np.multiply(1-np.multiply(kalman_gain,h),p)) + sigma_w_sq + \
    kalman_gain*v_resid*kalman_gain
    beta_list.append(beta)
```

```python
for j in range(10):
    plt.plot(stock_data_2.index, [i[j] for i in beta_list], label = stock_data_2.columns[j])
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.legend(loc="upper left");
```

```python
# average of Kalman filter betas

lis = [i[0] for i in beta_list]
print(sum(lis)/len(lis))
```

We can see that the beta value from CAPM is similar to the average beta value from Kalman filtering. The static CAPM uses the same value of $\beta$ throughout the period, as opposed to time-varying betas in the Dynamic Linear Model. The changing $\beta$ throughout the period gives a more accurate representation of the stock's risk. Ideally, we would not like to change the $\beta$ at every instant either and would like to strike a balance for constant $\beta$ values for smaller intervals which can then change as the interval changes.

```python
# plotting residuals for Apple

mkt_ret = mkt_data_2.loc[:,'sp500']
rf = mkt_data_2.loc[:,'3mTCM']
erp = (mkt_ret - rf).tolist()
stock_ret = stock_data_2.iloc[:,0]
excess_ret = (stock_ret - rf).tolist()
residuals_capm = [a-b for a,b in zip(excess_ret, [i*beta2[0] for i in erp])]

beta_kf_appl = [i[0] for i in beta_list]
residuals_kf = [a-b for a,b in zip(excess_ret, [i*j for i,j in zip(erp,beta_kf_appl)])]
```

```python
plt.plot(stock_data_2.index, residuals_capm);
plt.plot(stock_data_2.index, residuals_kf);
```
